```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import drive
```

```python
# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/DSpj01/creditcard.csv')
df.head()

```

```python
# Datatypes in DF
print(df.dtypes)
```

```python
# Calculating null values
df.isnull().sum()
```

```python
# Calculating Legitimate & Fraudulent values
fraud_counts = df['Class'].value_counts()
print(f"Legitimate: {fraud_counts[0]}, Fraudulent:{fraud_counts[1]}")
```

```python
# Percentage of Fraud Transactions
fraud_percentage = (fraud_counts[1] / len(df)) * 100
print(f"Percentage of Fraudulent Transactions: {fraud_percentage:.2f}%")
```

```python
# Min , Max, Mean and Median values for Amount
df['Amount'].describe()
```

```python
# Finding row with maximum transaction amount
max_amount = df[df['Amount'] == df['Amount'].max()]

# Extracting the amount and class
amount = max_amount['Amount'].values[0]
transaction_class = max_amount['Class'].values[0]

print(f"Amount      Class\n{amount:<12}{transaction_class}")
```

```python
# Creating bar chart : count of fraudulent vs legitimate transactions
sns.barplot(x=fraud_counts.index, y=fraud_counts.values, color="blue")
plt.title("Count of Fraudulent vs Legitimate Transactions")
plt.xlabel("Transaction Type (0: Legitimate, 1: Fraudulent)")
plt.ylabel("Count")
plt.show()
```

```python
# Creating histogram of transaction amount
plt.hist(df['Amount'], bins=50, color='blue', edgecolor='black')
plt.title("Histogram of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# Using heatmap for visualizing correlation between numerical features
plt.figure(figsize=(15, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='hot', linewidths=0.5)
plt.title("Correlation Heat Map ")
plt.show()
```
